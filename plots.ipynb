```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('dataset_winecheese.csv')

df_grouped = df.groupby(['tipo-vinho', 'tipo-combinacao']).size().unstack(fill_value=10)

# Plot stacked bar
fig, ax = plt.subplots(figsize=(20, 5))

# Add stacked bars
bottoms = np.zeros(len(df_grouped))
for category in df_grouped.columns:
    ax.bar(df_grouped.index, df_grouped[category], label=category, bottom=bottoms)
    bottoms += df_grouped[category]

ax.set_xlabel('Tipo de Vinho')
ax.set_ylabel('Quantidade de Combinaes')
ax.set_title('Combinaes de Vinho e Queijo')

ax.legend(title='Tipo de Combinao')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('dataset_winecheese.csv')

sns.set_theme(style="darkgrid")

height = 5
aspect = 1.5

g = sns.FacetGrid(df, col="tipo-combinacao", hue="tipo-vinho", col_wrap=4, palette='crest', height=height, aspect=aspect)
g.map(sns.scatterplot, "tipo-vinho", "tipo-queijo")

for ax in g.axes.flat:
     
    ticks = ax.get_xticks()
    labels = [label.get_text() for label in ax.get_xticklabels()]
    
    new_labels = [label.replace(' ', '\n') for label in labels]
    ax.set_xticks(ticks)
    ax.set_xticklabels(new_labels)

    ax.set_xlabel('')
    ax.set_ylabel('')

plt.show()
```
