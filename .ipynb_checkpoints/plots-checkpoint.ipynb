```python
import pandas as pd
import matplotlib.pyplot as plt

# Carrega o CSV em um DataFrame
df = pd.read_csv('dataset_winecheese.csv')

# Plota os dados
plt.plot(df['tipo-vinho'], df['tipo-queijo'])
plt.title('')
plt.xlabel('Tipo de vinho')
plt.ylabel('Tipo de queijo')
plt.show()
```

```python
pd.crosstab(df['tipo-queijo'], df['tipo-vinho']).plot(kind='bar', stacked=True)
plt.title('Distribuio de Tipos de Vinho por Tipo de Queijo')
plt.xlabel('Tipo de Queijo')
plt.ylabel('Quantidade')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('dataset_winecheese.csv')

sns.set_theme(style="darkgrid")
sns.countplot(x='tipo-vinho', data=df)
plt.title('Contagem de Tipos de Vinho')
plt.xlabel('Tipo de Vinho')
plt.ylabel('Quantidade')
plt.show()
```

```python
heatmap_data = pd.crosstab(df['tipo-vinho'], df['tipo-queijo'])
sns.heatmap(heatmap_data, annot=True, cmap='Purples')
plt.title('Frequncia de Combinaes de Vinho e Queijo')
plt.xlabel('Tipo de Queijo')
plt.ylabel('Tipo de vinho')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carrega o seu dataset
df = pd.read_csv('dataset_winecheese.csv')

df_sorted = df.sort_values("tipo-combinacao", ascending=True)

Configura o estilo do grfico
sns.set_theme(style="whitegrid")

palette = ["#FF6347", "#4682B4", "#32CD32"]
sns.set_palette(palette)

sns.barplot(x="tipo-combinacao", y="tipo-vinho", data=df_sorted, color="#FF6347")
sns.barplot(x="tipo-combinacao", y="tipo-queijo", data=df_sorted, color="#4682B4")

plt.xlabel('Tipo de combinao')
plt.ylabel('Tipo de Vinho / Queijo')
plt.title('Combinaes de Vinho e Queijo')

plt.show()
```
